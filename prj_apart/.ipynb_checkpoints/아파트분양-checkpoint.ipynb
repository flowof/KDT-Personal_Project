```python
df_last = pd.read_csv('data/주택도시보증공사_전국 평균 분양가격(2019년 12월).csv', encoding='cp949')
df_last.head()
```

```python
df_last.info()
```

```python
# isna를 통해 결측치를 구하자.
df_last.isnull().sum().to_frame()
#print('->분양가격(㎡)에 결측치 존재')
```

```python
# 분양가격이 object(문자) => 수치형으루 변경하자.
print(df_last.info())
print('-'*50)
print(df_last.head())
print('-'*50)
print(np.dtype(df_last['분양가격(㎡)']))
#print(type(df_last['분양가격(㎡)']))
```

```python
df_last['분양가격'] = pd.to_numeric(df_last['분양가격(㎡)'], errors='coerce')
print(df_last.head())
print('-'*50)
print( ' ->', np.dtype(df_last['분양가격(㎡)']), '에서', np.dtype(df_last['분양가격']), '으로 타입 변경됨')
```

```python
# 평당분양가격 구하기
df_last['평당분양가격'] = df_last['분양가격']*3.3
```

```python
print(df_last['규모구분'].unique())
print(np.dtype(df_last['규모구분']))
```

```python
# str 안에 있는 replace 해준다는 느낌으로 str. 붙여줌
df_last['전용면적'] = df_last['규모구분'].str.replace('전용면적', '')
df_last['전용면적'] = df_last['전용면적'].str.replace('초과', '~')
df_last['전용면적'] = df_last['전용면적'].str.replace('이하', '')
df_last['전용면적'] = df_last['전용면적'].str.replace(' ', '').str.strip()

df_last.head()
```

```python
df_last.info()
```

```python
df_last = df_last.drop(['규모구분', '분양가격(㎡)'], axis=1) # 컬럼 지움
```
