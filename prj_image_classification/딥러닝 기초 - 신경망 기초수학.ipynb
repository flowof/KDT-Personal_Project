## 신경망 기초 수학

```python
import math
import numpy as np
import matplotlib.pyplot as plt
plt.style.use(['seaborn-whitegrid'])
```

- y = ax + b
    - a:기울기, b:y절편

- 그래프 상에서 직선인 그래프

```python
def linear_function(x):
    a=0.5
    b=2
    
    return a*x + b # 0.5*x + 2
```

```python
print(linear_function(5))
```

```python
x = np.arange(-5, 5, 0.1)
y = linear_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Linear Function')
plt.show()
```

이차함수
- 일반적으로 두 개의 실근을 가짐
- y = ax^2 + bx + c

```python
def quadratic_function(x):
    a = 1
    b = -1
    c = -2
    
    return a*x**2 + b*x + c
```

```python
print(quadratic_function(2))
```

```python
x = np.arange(-5, 5, 0.1)
y = quadratic_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Quadratic Function')
plt.show()
```

삼차함수(다항함수)
- y = ax^3 + bx^2 + cx + d

```python
def cubic_function(x):
    a = 4
    b = 0
    c = -1
    d = -8
    
    return a*x**3 + b*x**2 + c*x + d # 4*x**3 + 0*x**2 + (-1)*x + d
```

```python
print(cubic_function(3))
```

```python
x = np.arange(-5, 5, 0.1)
y = cubic_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.title('Cubic Function')
plt.show()
```

함수의 최소값/최대값

```python
def my_func(x):
    a = 1
    b = -3
    c = 10
    return a*x**2 + b*x + c
```

```python
x = np.arange(-10, 10, 0.1)
y =my_func(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.scatter(1.5, my_func(1.5))
plt.text(1.5-1.5, my_func(1.5) + 10, 'min value of f(x)\n({},{})'.format(1.5, my_func(1.5)), fontsize=10)
plt.title('my_func')
plt.show()
```

```python
min_val = min(y)
print(min_val)
```

특정 구간 내에서 최소값 구하기

```python
def get_minimum(x1, x2, f):
    x = np.arange(x1, x2, 0.01)
    y = f(x)
    
    plt.plot(x,y)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.title('get_minimum')
    plt.show()
    
    return min(y)

print(get_minimum(1, 4, my_func))
```

지수함수/로그함수
- 지수함수-로그함수는 역함수 관계(y = x 직선 대칭 단, 밑이 같을 때)
- 파이썬으로 직접 구현 가능

지수함수
- y = a^x (a는 0이 아닐 때)(기본형)
- y = e^x (e = 2.71828...)

```python
def exponential_function(x):
    a = 4
    return a**x

print(exponential_function(4))
print(exponential_function(0))
```

```python
x = np.arange(-3, 2, 0.1)
y = exponential_function(x)

plt.plot(x,y)
plt.xlabel('x')
plt.ylabel('y')
plt.ylim(-1, 15)
plt.xlim(-4, 3)
plt.title('exponential_function')
plt.show()
```

```python
def exponential_function2(x):
    a = 4
    return math.pow(a, x)

print(exponential_function2(4))
print(exponential_function2(0))
```

밑이 e인 지수함수 표현

```python
print(math.exp(4))
print(np.exp(4))
```

로그함수

```python
print(math.log(2, 3)) # log(지수, 밑)?
print(np.log2(4)) # 밑이 2인 로그함수
print(np.log(4)) # 밑이 e인 자연로그
```
